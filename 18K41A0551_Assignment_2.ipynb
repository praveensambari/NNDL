```python
import sympy as sym
from sympy import *
#step 1
X=1
Y=2
n=0.1
epochs=10
itr = 1
x = sym.Symbol('x')
y = sym.Symbol('y')
f = x**2+y**2+10

print(f"STEP 1 : x={X},y={Y},epoch={epochs},n={n},itr={itr}, Function={f} \n")

while True:
  print(f"Iteration {itr}\n")
  #step 2
  dfx = sym.diff(f,x) 
  dfy = sym.diff(f,y) 
  val_x = dfx.subs(x,X)
  val_y = dfy.subs(y,Y)
  print(f"STEP 2 : First Order derivative df/dx at {round(X,2)} = {round(val_x,2)}\n\t First Order derivative df/dy at {round(Y,2)} = {round(val_y,2)}")
  #step 3
  delta_X = -(n)*val_x
  delta_Y = -(n)*val_y
  print(f"STEP 3 : Calaculate change in x,y\n\t delta x = {round(delta_X,2)}\n\t  delta y = {round(delta_Y,2)}")
  #step 4: 
  X = X+delta_X
  Y = Y+delta_Y
  print(f"STEP 4 : Update variables x,y\n\t x = {round(X,2)} \n\t y = {round(Y,2)} ")
  resf = f.subs(x,X)
  res = resf.subs(y,Y)
  #step 5
  print("STEP 5 : Update iteration\n")
  itr = itr+1
  #step 6
  print(f"STEP 6 : {itr} > {epochs} then ")
  if(itr>epochs):
    break;
  else:
    print("\tGoto STEP 2\n")
#step 7
print("STEP 7 : Global Minima point obtained at: ")
print("X :",round(X,2))
print("Y :",round(Y,2))
optimalres = f.subs(x,X)
optimalfinal = optimalres.subs(y,Y)
print("Optimal function value: ",round(optimalfinal,2))
```
